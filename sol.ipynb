```python
# 5. 	Plot top 5 sales for each Region

df = df.sort_values(by=['Region', 'Sales'], ascending=[True, False])

topFiveSales = df.groupby('Region').head(5)

plt.figure(figsize=(10, 6))
for region, data in topFiveSales.groupby('Region'):
    plt.bar(data['Name'], data['Sales'], label=region)

plt.xlabel('Name')
plt.ylabel('Sales')
plt.title('Top 5 Sales for Each Region')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# 6	Plot the chart -> percentage contribution of sales for Unique Name

namedSales = df.groupby('Name')['Sales'].sum().reset_index()
namedSales

total_sales = namedSales['Sales'].sum()
namedSales['Percentage'] = (namedSales['Sales'] / total_sales) * 100
namedSales['Percentage']
namedSales = namedSales.sort_values(by='Percentage', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(namedSales['Name'], namedSales['Percentage'])
plt.xlabel('Name')
plt.ylabel('Percentage Contribution (%)')
plt.title('Percentage Contribution of Sales for Unique Names')
plt.xticks(rotation=45)
plt.show()
```
